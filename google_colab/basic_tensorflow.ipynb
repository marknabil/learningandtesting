```python
import tensorflow as tf
```

```python
sess = tf.Session()
```

# Rank

```python
# rank 0 tensor: a scalar
mammal = tf.Variable('Elephant', tf.string)
ignition = tf.Variable(123, tf.int16)
floating = tf.Variable(7.7123456789, tf.float64)

print('rank of string scalar ', sess.run(tf.rank(mammal)))
print('rank of int scalar ', sess.run(tf.rank(ignition)))
print('rank of float scalar', sess.run(tf.rank(floating)))
```

```python
# rank 1 tensor: vector 
complex_number = tf.Variable((12.3, -4.85), tf.complex64)
str_value = tf.Variable(['Hello'], tf.string)
float_numbers = tf.Variable([1.23, 3.45, 6.78], tf.float32)
int_numbers = tf.Variable([1, 2, 3, 4, 5, 6, 7], tf.int32)

print('rank of complex numner ', sess.run(tf.rank(complex_number)))
print('rank of string vector ', sess.run(tf.rank(str_value)))
print('rank of float vector', sess.run(tf.rank(float_numbers)))
print('rank of int vector ', sess.run(tf.rank(int_numbers)))
```

```python
# rank 2 tensor: matrix
complex_numbers = tf.Variable([(1.23, 4.56), (7.54, 8.34)], tf.complex64)
int_matrix = tf.Variable([[1], [2]], tf.int16)
myxor = tf.Variable([[False, True], [True, False]], tf.bool)
linear_square = tf.Variable([[2], [9], [12], [25], [36]], tf.int32)
squarish_squares = tf.Variable([[4, 9], [16, 25]], tf.int32)

print('rank of complex number vector ', sess.run(tf.rank(complex_numbers)))
print('rank of int matrix ', sess.run(tf.rank(int_matrix)))
print('rank of boolean matrix ', sess.run(tf.rank(myxor)))
print('rank of int matrix ', sess.run(tf.rank(linear_square)))
print('rank of int matrix ', sess.run(tf.rank(squarish_squares)))
```

```python
# n rank tensor 
# example of 10 images, 299x299 with 3 color channel 
image_list = tf.zeros([10, 299, 299, 3])
second_image_list = tf.zeros(sess.run(tf.shape(image_list)))

print('rank of 4 dimension tensor ', sess.run(tf.rank(image_list)))
print('shape of tensor ', image_list.shape)
print('shape of second image list copy from image_list ', second_image_list.shape)
```

# Reshape

```python
 # 60 elements 3-dimension 
rank_three_tensor = tf.ones([3, 4, 5])

# 60 elements matrix 
matrix = tf.reshape(rank_three_tensor, [6, 10])

# -1 to tell reshape to calculate the size of this demension 
# the result still 60 elements matrix 
matrix_3_row = tf.reshape(matrix, [3, -1])
matrix_4_3_5 = tf.reshape(matrix_3_row, [4, 3, -1])

print('original tensor shape ', rank_three_tensor.shape)
print('reshape to matrix ', matrix.shape)
print('reshape to 3 row matrix ', matrix_3_row.shape)
print('reshape to 4x3x5 tensor ', matrix_4_3_5.shape)
```

```python
def print_parameters():
    W_val, b_val, loss_val = sess.run([W, b, loss], feed_dict={x: x_train, y: y_train})
    print('W: %s,  b: %s, loss: %s ' %(W_val, b_val, loss_val))
    print()

# training data 
x_train = [1, 2, 3, 4]
y_train = [0, -1, -2, -3]

# model parameters 
W = tf.Variable([.3], dtype=tf.float32)
b = tf.Variable([-.3], dtype=tf.float32)

# model input and output 
x = tf.placeholder(tf.float32)
y = tf.placeholder(tf.float32)
linear_model = W*x + b 

# loss 
loss = tf.reduce_sum(tf.square(linear_model - y))

# optimizer 
optimizer = tf.train.GradientDescentOptimizer(0.01)
train = optimizer.minimize(loss)

init = tf.global_variables_initializer()
sess = tf.Session()
sess.run(init)

print('Linear model output:')
print(sess.run(linear_model, feed_dict={x: x_train}))
print_parameters()

# re-assign value for W and b 
fixW = tf.assign(W, [-1.])
fixb = tf.assign(b, [1.])
sess.run([fixW, fixb])

print_parameters()

# reset values 
sess.run(init)
print('Before train:')
print_parameters()

# training loop 
for i in range(1000):
    sess.run(train, feed_dict={x: x_train, y: y_train})

print('Trained: ')
print_parameters()
```
